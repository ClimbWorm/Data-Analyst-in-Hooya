```python
import pandas as pd
import numpy as np
```

```python
#读取excel数据
#定义路径IO
IO = r'C:\Users\lirong1\Desktop\会员日复盘\data_prime_day.xlsx'
spider = pd.read_excel(io = IO,sheet_name=0,header=0,index_col= 0)
spider[:5]
```

```python
promising_asin = pd.read_excel(io = IO,sheet_name=1,header=0,index_col= 0)
promising_asin[:5]
```

```python
##转换数据格式,转换为数值型后面使用聚合函数就不会报错了
# 归类不同字段类型
spider_fieldtype_str = ['shop_name','category','sub_category']
spider_fieldtype_int = ['ask_num','category_rank','sub_category_rank']
spider_fieldtype_float = ['coupon','price','star','review']
spider_fieldtype_time = ['create_time']

asin_fieldtype_str = ['shop_name']
asin_fieldtype_int = ['impression','click','spend','sales']
asin_fieldtype_float = ['performance']

# 清洗spider表的数据类型
for i in spider_fieldtype_str:
    spider['{}'.format(i)] = spider['{}'.format(i)].astype(str)
    

for i in spider_fieldtype_int:
    spider['{}'.format(i)] = spider['{}'.format(i)].astype(int)
    
    
for i in spider_fieldtype_float:
    spider['{}'.format(i)] = spider['{}'.format(i)].astype(float)
    
    
for i in spider_fieldtype_time:
    spider['{}'.format(i)] = pd.to_datetime(spider['{}'.format(i)])
    
    
#清洗promising_asin表的数据类型    
for i in asin_fieldtype_str:
    promising_asin['{}'.format(i)] = promising_asin['{}'.format(i)].astype(str)
    
    
for i in asin_fieldtype_int:
    promising_asin['{}'.format(i)] = promising_asin['{}'.format(i)].astype(int)
    
    
for i in asin_fieldtype_float:
    promising_asin['{}'.format(i)] = promising_asin['{}'.format(i)].astype(float)
    
    
```

```python
#处理spider数据表,group by后求平均
spider_g = spider.groupby([spider.index,'category','sub_category'])['review','ask_num','category_rank','sub_category_rank','coupon','price','star'].mean()
spider_g[:5]
```

```python
##写码过程中用来测试的,不用管
##promising_asin.index.get_level_values('asin')[0]
#t1 = spider_g.iloc[spider_g.index.get_level_values('asin') == 'B00NCOZ9U2'].index.get_level_values('category')[0]
#t2 = spider_g.iloc[spider_g.index.get_level_values('asin') == 'B00NCOZ9U2'].index.get_level_values('sub_category')[0]
#spider_g.loc[('B00NCOZ9U2',t1,t2),'sub_category_rank']
#promising_asin.index.get_level_values('asin')
#perfect_table = pd.DataFrame(columns=['asin','shop_name', 'performance', 'impression','click','spend','sales','price','coupon','category','category_rank','sub_category','sub_category_rank','review','star','ask_num'])
```

```python
##构建最终的完整大表
perfect_table = pd.DataFrame(columns=['asin','shop_name', 'performance', 'impression','click','spend','sales','price','coupon','category','category_rank','sub_category','sub_category_rank','review','star','ask_num'])
manual_needed = np.setdiff1d(promising_asin.index,spider_g.index.get_level_values('asin'))#asin在promising_asin中有在spider_g(爬虫数据库)中却没有的
for i in promising_asin.index.get_level_values('asin'):
    temp = {}
    if i in manual_needed:
        temp['asin'] = i
        temp['shop_name'] = promising_asin.loc[i]['shop_name']
        temp['performance'] = promising_asin.loc[i]['performance']
        temp['impression'] = promising_asin.loc[i]['impression']
        temp['click'] = promising_asin.loc[i]['click']
        temp['spend'] = promising_asin.loc[i]['spend']
        temp['sales'] = promising_asin.loc[i]['sales']
        perfect_table = perfect_table.append(temp,ignore_index = True)
        continue
    else:
        category = spider_g.iloc[spider_g.index.get_level_values('asin') == i].index.get_level_values('category')[0]
        sub_category = spider_g.iloc[spider_g.index.get_level_values('asin') == i].index.get_level_values('sub_category')[0]
        temp['asin'] = i
        temp['shop_name'] = promising_asin.loc[i]['shop_name']
        temp['performance'] = promising_asin.loc[i]['performance']
        temp['impression'] = promising_asin.loc[i]['impression']
        temp['click'] = promising_asin.loc[i]['click']
        temp['spend'] = promising_asin.loc[i]['spend']
        temp['sales'] = promising_asin.loc[i]['sales']
        temp['price'] = spider_g.loc[(i,category,sub_category),'price']
        temp['coupon'] = spider_g.loc[(i,category,sub_category),'coupon']
        temp['category'] = category
        temp['category_rank'] = spider_g.loc[(i,category,sub_category),'category_rank']
        temp['sub_category'] = sub_category
        temp['sub_category_rank'] = spider_g.loc[(i,category,sub_category),'sub_category_rank']
        temp['review'] = spider_g.loc[(i,category,sub_category),'review']
        temp['star'] = spider_g.loc[(i,category,sub_category),'star']
        temp['ask_num'] = spider_g.loc[(i,category,sub_category),'ask_num']
        perfect_table = perfect_table.append(temp,ignore_index = True)
print(manual_needed)
```

```python
#将整合完成的表格写入csv中
with open('perfect_table.csv', 'a', encoding='utf-8',newline="") as f:
        perfect_table.to_csv(f, mode='a+', index=False,  header = True)

perfect_table_result = pd.read_csv('perfect_table.csv')
print(perfect_table_result)
# 关闭文件
f.close()#只有在文件关闭状态下才不会报错
```
